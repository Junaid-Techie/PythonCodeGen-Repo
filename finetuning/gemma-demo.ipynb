```python
from llama_cpp import Llama
from typing import Optional
import time
from huggingface_hub import hf_hub_download

def generate_prompt(input_text):
    text = f"<start_of_turn>user\n{input_text}<end_of_turn>\n<start_of_turn>model\n"
    return text

# Set up the parameters
repo_id = "vdpappu/gemma2_coding_assistant_gguf"
filename = "gemma2_coding.gguf"
local_dir = "."

downloaded_file_path = hf_hub_download(repo_id=repo_id, filename=filename, local_dir=local_dir)
print(f"File downloaded to: {downloaded_file_path}")
```

```python
# Load the model 
llm = Llama(model_path=downloaded_file_path, verbose=False)
question = "Generate a Python code for crawling a website for a specific type of data."
prompt = generate_prompt(input_text=question)

output = llm(prompt, 
             temperature=0.7,
             top_p=0.9,
             top_k=50,
             repeat_penalty=1.5,
             max_tokens=200, 
             stop=["Question:","<eos>"])
print(output['choices'][0]['text'])
```
